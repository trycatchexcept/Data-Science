```python
luxury_homes = df[df['bathrooms'] > 3.75]
luxury_homes.index
```

```python
df_no_luxury = df.drop(luxury_homes.index)
```

```python
sns.boxplot(df_no_luxury['bedrooms'])
```

```python
df_no_luxury[df_no_luxury['bedrooms'] > 5].describe()
```

```python
from scipy import stats
```

```python
# Winsorization
df_no_luxury["bedrooms_winsorized"] = stats.mstats.winsorize(df_no_luxury['bedrooms'], limits=[0.01,0.01])
```

```python
df_no_luxury[['bedrooms','bedrooms_winsorized']]
```

```python
df_no_luxury.describe()
```

```python
# Winsorization
df_no_luxury["sqft_living_winsorized"] = stats.mstats.winsorize(df_no_luxury['sqft_living'], limits=[0.01,0.01])
```

```python
df_no_luxury[["sqft_living","sqft_living_winsorized"]]
```

```python
df_no_luxury.drop('id',axis=1,inplace=True)
df_no_luxury.head()
```

```python
# Extracting values from dates
df_no_luxury['sale_month'] = df_no_luxury['date'].apply(lambda x : x.month)
df_no_luxury
```
